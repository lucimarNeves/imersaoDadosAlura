Criar um gr√°fico de barras com informa√ß√µes de composto

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8,6))
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos').sort_values(by = "composto", ascending = False), 
                  order=cod_compostos, palette="Blues_d")
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
cod_compostos = dados['composto'].value_counts().sort_values(ascending=False).index[:5]
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=dados.query('composto in @cod_compostos'), x='composto', order=cod_compostos, palette="Blues_d")
ax.set_title("Top 5 compostos", fontweight='bold', fontsize=20)
ax.set_xlabel('Compostos',fontweight='bold')
ax.set_ylabel('Frequ√™ncia',fontweight='bold')
plt.show()
```

Correla√ß√£o entre as vari√°veis

```python
dados.loc[:,'g-0':'g-77'].corr()
```

```python
corr = dados.loc[:,'g-0':'g-50'].corr()
```

```python
import numpy as np


# Compute the correlation matrix

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
sns.set()
```

```python
corr_celular = dados.loc[:,'c-0':'c-50'].corr()
```

```python
# Compute the correlation matrix

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
sns.set()
```
